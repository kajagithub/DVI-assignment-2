```python
# Correlation matrix (CM) for observation purposes

col_of_interest = ['highway-mpg', 'engine-size', 'peak-rpm', 'price']

#Correlation heat map
correlation_matrix = auto[col_of_interest].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", linewidths=0.5, cbar=True).set(title='Correlation Heatmap')
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Set colors and titles for each subplot
colors = ['#33FF57','#FF5733', '#5733FF']
titles = ['Regression Plot for Engine Size vs. Price',
          'Regression Plot for Highway MPG vs. Price',
          'Regression Plot for Peak RPM vs. Price']

# Create regression plots for each variable pair
for i, (x_var, title) in enumerate(zip(['engine-size', 'highway-mpg', 'peak-rpm'], titles)):
    sns.regplot(data=auto, x=x_var, y='price', ax=axes[i], color=colors[i])
    axes[i].set_title(title, fontsize=14, fontweight='bold')
    axes[i].set_xlabel(x_var.replace('-', ' ').title(), fontsize=12)
    axes[i].set_ylabel('Price', fontsize=12)
    axes[i].tick_params(axis='both', labelsize=10)

# Show the plots
plt.tight_layout()
plt.show()
```

**Observations**:

1. **Engine Size vs. Price**:
   - Observation: There appears to be a positive linear relationship between "engine-size" and "price." As "engine-size" increases, "price" tends to increase.
   - Potential Predictor: "Engine-size" is a potential predictor variable of "price."

2. **Highway MPG vs. Price**:
   - Observation: There is a negative linear relationship between "highway-mpg" and "price." As "highway-mpg" increases (indicating better fuel efficiency), "price" tends to decrease.
   - Potential Predictor: "Highway-mpg" is a potential predictor variable of "price."

3. **Peak RPM vs. Price**:
   - Observation: There is no strong linear relationship between "peak-rpm" and "price." The data points are scattered without a clear linear pattern.
   - Potential Predictor: "Peak-rpm" is not a strong predictor variable of "price."

**Pre-attentive Attributes Used**:

2. **Position (X and Y)**: The position of each data point on the X and Y axes represents the values of the corresponding variables, making it easy to visually compare and assess the relationship between variables.

**Gestalt Principles Used**:

1. **Proximity**: In each scatter plot, the proximity of data points to each other on the chart conveys their relationship. Data points that are closer to each other tend to be related in terms of their "price" and the respective variable being analyzed.

2. **Similarity**: The similarity in the appearance of data points (e.g., size, shape) in the scatter plots aids in grouping and identifying patterns or trends in the data.

In summary, based on the visualizations and analysis:

```python
###computing basic statistics###
#Extracting continuous variables from data set
continuous_vars = auto.select_dtypes(include=['float64', 'int64'])
continuous_var_list = continuous_vars.columns.tolist()

import scipy.stats as stats

# Function to calculate 5-point summary
def five_point_summary(data):
    return [
        data.min(),
        np.percentile(data, 25),
        np.percentile(data, 50),
        np.percentile(data, 75),
        data.max()
    ]

# Function to calculate IQR
def iqr(data):
    return np.percentile(data, 75) - np.percentile(data, 25)

# Calculate skewness and kurtosis
skewness = continuous_vars.apply(stats.skew)
kurt = continuous_vars.apply(stats.kurtosis)

# Create a summary table
summary_table = pd.DataFrame({
    'Min': continuous_vars.min(),
    'Q1': continuous_vars.apply(lambda x: np.percentile(x, 25)),
    'Median (Q2)': continuous_vars.median(),
    'Q3': continuous_vars.apply(lambda x: np.percentile(x, 75)),
    'Max': continuous_vars.max(),
    'IQR': continuous_vars.apply(iqr),
    'Skewness': skewness,
    'Kurtosis': kurt
})

# Print the 5-point summary, skewness, and kurtosis
print("5-Point Summary, Skewness, and Kurtosis:")
print(summary_table)

```

```python
#List all continuous variable attributes

print("continuous variables: ", len(continuous_var_list))
print(continuous_var_list)
print("")

# Create a boxplot for each continuous variable
fig, axes = plt.subplots(4, 5, figsize=(16, 16))
fig.suptitle("Descriptive Statistical Analysis of Continuous Variables", fontsize=16, fontweight='bold')
fig.subplots_adjust(wspace=0.5, hspace=0.5)
for i, var in enumerate(continuous_vars):
    row, col = divmod(i, 5)
    sns.boxplot(data=auto, y=var, ax=axes[row, col])
    axes[row, col].set_title(f'{var}')
    axes[row, col].set_xlabel('')
    axes[row, col].set_ylabel('')

# Remove empty subplots
for i in range(len(continuous_vars), 4 * 5):
    row, col = divmod(i, 5)
    fig.delaxes(axes[row, col])

# Show the boxplots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

<h4>Perform the data preprocessing required.<br>

<p>The heatmap should plot  the target variable (price) proportional to colour with respect to the variables 'drive-wheel' and 'body-style' in the vertical and horizontal axis respectively. This allows us to visualize how the price is related to 'drive-wheel' and 'body-style'.</p>

```python
# Pivot the data to create the required format for the heatmap
heatmap_data = auto.pivot_table(index='body-style', columns='drive-wheels', values='price')

# Normalize the price values for better visualization
heatmap_data = (heatmap_data - heatmap_data.min().min()) / (heatmap_data.max().max() - heatmap_data.min().min())

# Reorder rows and columns for better readability
heatmap_data = heatmap_data.reindex(index=['sedan', 'hatchback', 'wagon', 'hardtop', 'convertible'],
                                    columns=['fwd', 'rwd', '4wd'])

# Create the heatmap
plt.figure(figsize=(9, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt=".5f", cbar=True)
plt.title('Heatmap: Relationship between Body Style and Drive-wheel vs. Price')
plt.xlabel('Drive-wheel')
plt.ylabel('Body Style')

plt.show()
```

From the heatmap, we can draw several key inferences regarding the relationship between "Body Style" and "Drive-wheel" with respect to car prices:

1. **Price Differences by Body Style and Drive-wheel**: The intensity of the colors in the heatmap represents the normalized car prices. We can observe that "convertible" and "hardtop" body styles generally have higher prices compared to "sedan," "hatchback," and "wagon" body styles. Additionally, for most body styles, "rwd" (rear-wheel drive) vehicles tend to have higher prices than "fwd" (front-wheel drive) and "4wd" (four-wheel drive) vehicles.

2. **Convertible and Hardtop Premium**: Among all body styles, "convertible" and "hardtop" vehicles with "rwd" have the highest prices, indicated by the darkest blue shades. This suggests that these body styles, especially when combined with rear-wheel drive, command a premium price.

3. **Sedans and Hatchbacks**: "Sedans" and "hatchbacks" with "fwd" are generally more affordable, shown by the lighter blue shades. This may indicate that these configurations are more budget-friendly options.

4. **Variation Within Body Styles**: Within a particular body style category, there can still be price variations based on the drive-wheel configuration. For example, "rwd" sedans generally have higher prices than "fwd" sedans.

5. **4wd Observations**: There is limited data for "4wd" configurations, but they tend to be in the mid-price range among body styles.

6. **Normalized Prices**: Keep in mind that the prices are normalized, so the specific price values are not represented here. Instead, we are focusing on relative price differences.

Though this heatmap provides insights into the relationships between body style, drive-wheel, and price, it doesn't capture all factors that influence car prices. Other factors such as engine size, brand, and additional features can also play a significant role in determining the price of a car.


**Pre-attentive Attributes:**
1. **Color**: We used a single color gradient (from light to dark blue) to represent the price values, making it easy to distinguish higher and lower prices.
2. **Position**: The position of each cell within the heatmap grid represents the combination of "Body Style" and "Drive-wheel," allowing us to compare different categories.
3. **Size**: The size of each cell (square) is standardized, ensuring that the focus remains on the color gradient and not on variations in cell size.
4. **Annotation**: Numerical values (price) are annotated within each cell, providing exact values for reference.

**Gestalt Principles:**
1. **Proximity**: Cells with similar values (related body styles and drive wheels) are grouped together, aiding in the comparison of prices within those groups.
2. **Similarity**: The color gradient within each cell is similar to other cells, allowing us to quickly identify patterns and variations in prices.
3. **Closure**: The heatmap's grid structure forms a sense of closure, helping us perceive individual cells as data points within the whole.
4. **Continuity**: The smooth color gradient within each cell and across the heatmap grid provides continuity in visualizing price trends, making it easier to identify price patterns.
